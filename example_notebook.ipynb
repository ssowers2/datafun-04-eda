```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.axes import Axes
```

```python
# Step 3. Initial Data Inspection

# Load the Iris dataset into pandas DataFrame
iris_df: pd.DataFrame = sns.load_dataset('iris')

# Inspect the first 3 rows of the DataFrame
display(iris_df.head(3))

# Inspect the shape
print(f"Shape of the dataset: {iris_df.shape}")

# Inspect the data types
print("Data Types:\n", iris_df.dtypes)

# General DataFrame information
iris_df.info()

# Rename columns
iris_df.rename(columns={
    'sepal_length': 'Sepal Length',
    'sepal_width': 'Sepal Width',
    'petal_length': 'Petal Length',
    'petal_width': 'Petal Width',
    'species': 'Species',
}, inplace=True)

# Pairplot with species differentiation
sns.pairplot(iris_df, hue='Species')

# List column names
print("Column Names:", iris_df.columns.tolist())

# Step 4. Initial Descriptive Statistics
# Summary statistics for numerical and categorical columns
display(iris_df.describe(include="all"))

# Inspect histogram by one numerical column (using renamed column)
iris_df["Sepal Width"].hist()
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.show()

# Show all plots
plt.show()
```

**Insights:**  
The pairplot shows that **Setosa** is clearly separated from the other species, while **Versicolor** and **Virginica** overlap slightly but can be distinguished using **petal length** and **petal width**.
