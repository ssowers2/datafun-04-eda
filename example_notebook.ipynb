## 1. Imports

```python
# Imports

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from matplotlib.axes import Axes
```

## 2. Load Data

```python
# 2. Load Data
iris_df: pd.DataFrame = sns.load_dataset('iris')
```

## 3. Initial Data Inspection


```python
# Initial Data Inspection

# Inspect the first 3 rows of the DataFrame
display(iris_df.head(3))

# Inspect the shape
print(f"Shape of the dataset: {iris_df.shape}")

# Inspect the data types
print("Data Types:\n", iris_df.dtypes)

# General DataFrame information
iris_df.info()

# List column names
print("Column Names:", iris_df.columns.tolist())
```

# Observations:
- The dataset contains 150 rows and 5 columns.
- The columns include sepal length, sepal width, petal length, petal width, and species.
- The data types are as expected: float64 for numerical measurements and object for the species category

## 4. Initial Descriptive Statistics

```python
# Initial Descriptive Statistics. Summary statistics for numerical and categorical columns
display(iris_df.describe(include="all"))
```

## Observations:
- Sepal Width has a mean of around 3.06 cm with a slight spread, while Petal Length shows greater variability across species.
- The dataset is balanced, with 50 samples for each species.

## 5. Initial Data Distribution for Numerical Columns

```python
# Histogram for Sepal Width
iris_df["sepal_width"].hist()
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.title("Histogram of Sepal Width")
plt.show()
```

## Observations:
- Sepal Width shows a roughly normal distribution with a slight skew.

## 6. Initial Data Transformation and Feature Engineering

```python
# Renamed columns
iris_df.rename(columns={
    'sepal_length': 'Sepal Length',
    'sepal_width': 'Sepal Width',
    'petal_length': 'Petal Length',
    'petal_width': 'Petal Width',
    'species': 'Species',
}, inplace=True)

# Feature Engineering by adding a Sepal Area
iris_df['Sepal Area'] = iris_df['Sepal Length'] * iris_df['Sepal Width']
plt.show()
```

## 7. Initial Visualizations

```python
# Pairplot with species differentiation
sns.pairplot(iris_df, hue='Species')
plt.suptitle("Pairwise Relationships of Iris Features", y=1.02)
plt.show()
```

## Observations:
- Setosa is separated from the other species, while Versicolor and Virginica overlap slightly but can be distinguished using petal length and petal width.

```python
# Scatter plot of Sepal Length vs Sepal Area
scatter_plt: Axes = sns.scatterplot(data=iris_df, x="Sepal Length", y="Sepal Area", hue="Species")
scatter_plt.set_xlabel("Sepal Length (cm)")
scatter_plt.set_ylabel("Sepal Area (cm)")
scatter_plt.set_title("Iris Sepal Length vs. Sepal Area by Species")
plt.show()
```

## Observations:
- A positive correlation exists between Sepal Length and Sepal Area across species.

## 8. Initial Insights

- Setosa is easily identifiable due to its small petal dimensions.
- Petal Length and Petal Width are the most effective ##features for differentiating species.
- The dataset is balanced, making it suitable for classification tasks.
